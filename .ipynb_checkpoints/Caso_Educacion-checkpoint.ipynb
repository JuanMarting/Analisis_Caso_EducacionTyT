Procedemos a dar solución a la interrogante #1, para ello emplearemos la siguiente función para determinar el tamaño optimo de la muestra:

```python
"""
Esta función retorna el tamaño optimo para un muestreo segun:
tam_pobl = tamaño población
prop = proporcion de la muestra
error = error admisible
conf = nivel de confianza deseado

return n = al numero de muestra optimo
"""
def tam_muestra(tam_pobl,prop,error,conf):
    N=tam_pobl
    p=prop
    e=error
    z=norm.ppf(q=(1-conf)/2)
    n=(N*p*(1-p)*z**2)/(((N-1)*e**2)+(p*(1-p)*z**2))
    return round(n)
```

Los datos que tenemos suministrados son los siguientes:

*tamaño total de la población 52712

*margen de error 5%

*nivel de confianza del 95%

```python
tam_optimo = tam_muestra(52712,0.5,0.05,0.95)

print(f"A partir de los requerimientos solicitados, se determina el tamaño optimo de muestreo: {tam_optimo}")
```

Para reponder el interrogante 2, los datos de la variable(cualitativa); procedemos a realizar un analisis descriptivo a partir de la siguiente función:

```python
"""
Esta función devuelve una tabla con un analisis descriptivo completo de la columna suministrada
Args:
    columnas (variable cuantitativa)
    
"""
def tabla_descriptivas(columnas):
    tabla_descriptivas=pd.DataFrame(columnas.describe())
    tabla_descriptivas.loc['coef. variation']=columnas.std()/columnas.mean() # Coeficiente variación
    tabla_descriptivas.loc['skew']=skew(columnas) # coeficiente de asimetria
    tabla_descriptivas.loc['kurtosis']=kurtosis(columnas) # coeficiente de curtosis.
    return tabla_descriptivas
```

```python
descrip_razonamiento = tabla_descriptivas(data['MOD_RAZONA_CUANTITAT_PUNT'])
descrip_razonamiento
```

```python


# Graficamos el histograma
hist_rc= sns.histplot(x=data['MOD_RAZONA_CUANTITAT_PUNT'],kde=True,color="steelblue", bins = 15)
hist_rc.set_xlabel('Puntaje')
hist_rc.set_title('Distribución del puntaje obtenido en razonamiento cuantitativo')

plt.show()  
```

**Conclusión**: Se evidencia que en el centro de distribucion, los puntajes obtenidos en el modulo de razonamiento cuantitativo fueron de 98 para el 50% de los datos agrupados, alrededor del centro de este mismo, se aprecia que la mayor cantidad de datos se agrupan entre los 83 y 112 puntos. Los datos presentan una dispersion baja (de acuerdo a la escala propuesta por Vargas (2007)). Finalmente, es posible afirmar que la distribución es relativamente simetrica(leve sesgo a la derecha) y achatada(colas cortas con poca frecuencia de datos extremos).

Para el interrogante 3, para determinar la mejor estimación de la media del puntaje en razonamiento cuantitativo poblacional empezaremos por obtener medias muestrales de nuestra media:

```python
# Obtenemos 10 datos para crear nuestra media muestral M1
aleatorio = data["MOD_RAZONA_CUANTITAT_PUNT"].sample(n=10).reset_index(drop = True)
df = pd.DataFrame({'M1': aleatorio})
df
```

Emplearemos la siguiente función para hacer la creación de media muestral de forma mas dinamica:

```python
#Funcion para crear media muestral:
def media_muestral(df,columna):
    for i in range(2,len(columna)+1):
        name_m = 'M' +str(i)
        # obtengo 10 muestras aleatorias de mi columna, se permiten repetidos
        aleatorios = columna.sample(n=10).reset_index(drop = True)
        df_M = pd.DataFrame({name_m: aleatorios})
        df = pd.concat([df, df_M], axis=1)
    return df
```

```python
dfm_razonamiento =  media_muestral(df,data["MOD_RAZONA_CUANTITAT_PUNT"])
dfm_razonamiento
```

```python
dfm_razonamiento.mean()
```

```python
# Distribución de la media muestral razonamiento cuantitativo:
plt.figure(figsize= (8,6))
ax=sns.histplot(x=dfm_razonamiento.mean(),kde=True,color="steelblue") # histograma
ax.annotate('Media:'+ str(round(np.mean(dfm_razonamiento.mean()),1)),xy=(110,50)) # Para visualizar la 'media'
ax.set_title('Distribución Media Muestral del Puntaje Obtenido en Razonamiento Cuantitativo', color='r')
plt.show()
```

**Nota**: La dispersión de la distribución muestral de la muestra se reduce entre mayor sea el tamaño de muestra, comprobando asi la teoría de: a mayor tamaño de muestra, menor el error estándar de la media.

**Conclusion**: La mejor estimación es mediante la distribucion de la media muestral. Si se quiere un parametro fijo o preciso basta con aumentar el numero de valores tomados de forma aleatoria de la muestra de la población.

Para el punto 6, requerimos obtener la edad a partir de la columna "ESTU_FECHANACIMIENTO", entonces:

En el punto 9, procedemos a obtener la proporcion de hombres y mujeres que presentaron el icfes:

```python
## Función para construir tablas de frecuencias
def tabla_freq(columna):
    x=pd.crosstab(index=columna,columns="count").reset_index()
    x=x.rename(columns={'count':'Frec. Abs.'})
    y=pd.crosstab(index=columna,columns="count", normalize='columns').reset_index()
    y=y.rename(columns={'count':'Frec. Rel.'})
    tabla=x.merge(y,how='left')
    tabla.rename_axis(None, axis=1)
    return tabla
```

```python
tabla_genero = tabla_freq(data['ESTU_GENERO'])
tabla_genero
```

```python
plt.figure(figsize=(10,5)) 
plt.pie(tabla_genero["Frec. Rel."], labels=tabla_genero["ESTU_GENERO"],autopct='%1.1f%%',startangle=90)
plt.title("Proporción de Mujeres y Hombres ", color='r')
plt.show()
```

**Conclusión:** Según la proporción de la distribución de mujeres(F) y hombres(M). Se evidencia que la presentación de las pruebas icfes del año 2016 fue muy equilibrada en cuestion de la participación de mujeres como de hombres.

Para solucionar el punto 10, la mejor estimación de la proporcion poblacional de mujeres; se obtiene:

```python
# Obtenemos 10 datos para crear nuestra media muestral M1
datos_aleatorios = data["ESTU_GENERO"].sample(n=10).reset_index(drop = True)
df_gender = pd.DataFrame({'M1': datos_aleatorios})
df_gender
```

```python
muestral_gender = media_muestral(df_gender, data["ESTU_GENERO"])
muestral_gender
```

```python
# Función para calcular la proporción de 'F' en una columna
def calcular_proporcion(columna):
    return columna.value_counts(normalize=True).get('F', 0)
```

```python
# Calcular las proporciones de 'F' en cada columna
proporciones_f = muestral_gender.apply(calcular_proporcion) * 100 # obtenemos los porcentajes
proporciones_f
```

```python
# Distribución muestral de las proporciones estudiantes mujeres:
plt.figure(figsize= (8,6))
hist_prop_mujeres=sns.histplot(x=proporciones_f,kde=True,color="steelblue") # histograma
hist_prop_mujeres.annotate('Proporción muestral: '+ str(round(np.mean(proporciones_f),1)),xy=(65,100)) # Para visualizar la 'media'
hist_prop_mujeres.set_title('Distribución Muestral Proporción de  Mujeres Estudiantes', color= 'r')
plt.show()
```

**Conclusión:** La mejor estimación de la proporción de mujeres, se obtiene a traves la distribucion muestral de las proporciones tomando multiples muestras de los datos de forma aleatoria y calculando el promedio de la proporción, evidenceando que a mayor toma de proporciones muestrales(M) mas preciso sera el valor obtenido para la estimación de la población de mujeres estudiantes que presentaron las pruebas.

Del punto 12, tendremos en cuenta lo siguiente:
Un asesor considera que: 2 de cada 3 estudiantes, son mujeres.
Para este caso, se considera una proporción de 66.6% aproximadamente para las estudiantes mujeres

Ho: La proporción de estudiantes mujeres es del 66.6%.

Ha: La proporción de estudiantes mujeres **NO** es del 66.6%.

```python
## prueba z proporciones (número mujeres, total muestra, value=hipótesis de proporción poblacional)
z,vp = sm.proportion.proportions_ztest(186,381,value=0.666)
print(f'Estadístico Z= {z}, Valor-p= {vp}')
```

**Conclusión:** El valor-p es menor al nivel de significancia del 5%, se puede rechazar la hipótesis nula. Con un 95% de confianza podemos concluir que la proporcion de estudiantes mujeres **NO** es del 66.6%, Partiendo de lo anterior es correcto afirmar que el asesor esta errado.
